# Part I: Foundations

## 1. Sensitive Dependence on Initial Conditions

The hallmark of chaotic systems ‚Äî two trajectories starting infinitesimally close diverge exponentially:

$$\Large \boxed{|\delta \mathbf{x}(t)| \sim |\delta \mathbf{x}_0| \cdot e^{\lambda t}}$$

where $\lambda > 0$ is the **Lyapunov exponent** (measured in bits/time).

# Part II: The Van der Pol Oscillator

## 2. Governing Equation

$$\Large \boxed{\frac{d^2x}{dt^2} - \mu(1-x^2)\frac{dx}{dt} + x = 0}$$

<br>

### 2.1 First-Order System

Introducing velocity $y = \dot{x}$:

$$\begin{cases}
\dot{x} = y \\[8pt]
\dot{y} = \mu(1-x^2)y - x
\end{cases}$$

<br>

### 2.2 Damping Mechanism

| Condition | Damping Sign | Effect |
|:---------:|:------------:|:------:|
| $\|x\| < 1$ | **Negative** | Energy pumped IN |
| $\|x\| > 1$ | **Positive** | Energy dissipated OUT |

<br>

### 2.3 Jacobian Matrix

$$\mathbf{J}(x, y) = \begin{pmatrix}
0 & 1 \\[4pt]
-2\mu xy - 1 & \mu(1-x^2)
\end{pmatrix}$$

**At the origin** $(0, 0)$:

$$\mathbf{J}_0 = \begin{pmatrix}
0 & 1 \\[4pt]
-1 & \mu
\end{pmatrix}$$

<br>

### 2.4 Eigenvalue Analysis

**Characteristic equation:**

$$\lambda^2 - \mu\lambda + 1 = 0$$

**Solutions:**

$$\Large \lambda_{1,2} = \frac{\mu \pm \sqrt{\mu^2 - 4}}{2}$$

<br>

### 2.5 Stability Classification

| Parameter | Eigenvalues | Type | Behavior |
|:---------:|:-----------:|:----:|:--------:|
| $\mu = 0$ | $\pm i$ | Center | Harmonic oscillator |
| $0 < \mu < 2$ | Complex, Re$>$0 | Unstable spiral | Outward spiraling |
| $\mu \geq 2$ | Real, both $>$0 | Unstable node | Fast divergence |

**Conclusion:** For $\mu > 0$, origin is unstable ‚Üí **limit cycle exists** (Poincar√©-Bendixson)

<br>

### 2.6 Relaxation Oscillation Period

For large $\mu$:

$$\Large T \approx (3 - 2\ln 2)\mu \approx 1.614\,\mu$$

# Part III: The Lorenz System

## 3. Governing Equations

$$\Large \boxed{\begin{aligned}
\dot{x} &= \sigma(y - x) \\[6pt]
\dot{y} &= x(\rho - z) - y \\[6pt]
\dot{z} &= xy - \beta z
\end{aligned}}$$

<br>

### 3.1 Standard Parameters (Chaotic Regime)

| Symbol | Name | Physical Meaning | Value |
|:------:|:----:|:-----------------|:-----:|
| $\sigma$ | Prandtl number | Momentum / thermal diffusivity | **10** |
| $\rho$ | Rayleigh number | Temperature gradient strength | **28** |
| $\beta$ | Geometric factor | Cell aspect ratio | **8/3** |

<br>

---

## 4. Equilibrium Analysis

### 4.1 Finding Fixed Points

Setting all derivatives to zero:

$$\sigma(y - x) = 0 \quad \Rightarrow \quad y = x$$

$$x(\rho - z) - y = 0 \quad \Rightarrow \quad x(\rho - z - 1) = 0$$

$$xy - \beta z = 0$$

<br>

**Solution 1:** Trivial equilibrium (no convection)

$$\Large \boxed{O = (0, 0, 0)}$$

<br>

**Solution 2:** Non-trivial equilibria (for $\rho > 1$)

From $z = \rho - 1$ and $x^2 = \beta(\rho - 1)$:

$$\Large \boxed{C^{\pm} = \left(\pm\sqrt{\beta(\rho-1)},\; \pm\sqrt{\beta(\rho-1)},\; \rho-1\right)}$$

With standard parameters: $C^+ = (8.485, 8.485, 27)$

<br>

---

### 4.2 The Jacobian Matrix

$$\mathbf{J} = \begin{pmatrix}
-\sigma & \sigma & 0 \\[6pt]
\rho - z & -1 & -x \\[6pt]
y & x & -\beta
\end{pmatrix}$$

<br>

### 4.3 Stability of the Origin

$$\mathbf{J}_O = \begin{pmatrix}
-10 & 10 & 0 \\[4pt]
28 & -1 & 0 \\[4pt]
0 & 0 & -8/3
\end{pmatrix}$$

**Eigenvalues:**
- $\lambda_1 = -\beta = -2.667$ ‚úì
- $\lambda_2 \approx +11.83$ ‚úó
- $\lambda_3 \approx -22.83$ ‚úì

**Result:** Origin is a **saddle point** (2D stable, 1D unstable manifold)

### 4.4 Stability of $C^{\pm}$

At $C^+ = (8.485, 8.485, 27)$:

$$\mathbf{J}_{C^+} = \begin{pmatrix}
-10 & 10 & 0 \\[4pt]
1 & -1 & -8.485 \\[4pt]
8.485 & 8.485 & -2.667
\end{pmatrix}$$

<br>

**Characteristic polynomial:**

$$\lambda^3 + (\sigma + \beta + 1)\lambda^2 + (\sigma + \rho)\beta\,\lambda + 2\sigma\beta(\rho - 1) = 0$$

**Eigenvalues:**
- $\lambda_1 \approx -13.85$ (real, stable direction)
- $\lambda_{2,3} \approx 0.094 \pm 10.19i$ (complex, **unstable spiral!**)

# Part IV: The Hopf Bifurcation

## 5. Critical Rayleigh Number

The system transitions to chaos when $C^{\pm}$ lose stability:

$$\Large \boxed{\rho_H = \frac{\sigma(\sigma + \beta + 3)}{\sigma - \beta - 1} \approx 24.74}$$

<br>

### 5.1 Bifurcation Diagram

| $\rho$ Range | Equilibria | Behavior |
|:------------:|:----------:|:--------:|
| $\rho < 1$ | Only $O$ (stable) | No convection |
| $1 < \rho < 24.74$ | $O$ unstable, $C^{\pm}$ stable | Steady convection |
| $\rho > 24.74$ | All unstable | **Strange attractor** |

<br>

### 5.2 Lyapunov Exponent

At $\rho = 28$:

$$\Large \lambda_{\max} \approx 0.906 \text{ bits/time}$$

**Prediction horizon:**

$$T_{\text{pred}} \approx \frac{1}{\lambda_{\max}} \ln\left(\frac{\Delta_{\text{tol}}}{\Delta_0}\right)$$

### 6.2 OGY Control (Ott-Grebogi-Yorke)

# Part VI: Universal Route to Chaos

## 7. The Logistic Map

$$\Large x_{n+1} = r \cdot x_n \cdot (1 - x_n)$$

<br>

### 7.1 Period-Doubling Cascade

| Parameter $r$ | Dynamics |
|:-------------:|:--------:|
| $r < 1$ | Extinction ($x \to 0$) |
| $1 < r < 3$ | Stable fixed point |
| $r \approx 3.0$ | Period-2 bifurcation |
| $r \approx 3.45$ | Period-4 |
| $r \approx 3.54$ | Period-8, 16, 32... |
| $r > 3.57$ | **Chaos** |

<br>

### 7.2 Feigenbaum's Universal Constant

$$\Large \boxed{\delta = \lim_{n\to\infty} \frac{r_{n} - r_{n-1}}{r_{n+1} - r_n} = 4.669201609\ldots}$$

This constant appears in **all** systems undergoing period-doubling ‚Äî a remarkable universality!
